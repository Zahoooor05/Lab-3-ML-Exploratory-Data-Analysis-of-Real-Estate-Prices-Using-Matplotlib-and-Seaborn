```python
#Basic Line Plot: House Age vs House Price
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Real_Estate.csv")

x = df["House age"]
y = df["House price of unit area"]

plt.figure(figsize=(8,5))
plt.plot(x, y, 'b.')
plt.xlabel("House Age")
plt.ylabel("House Price of Unit Area")
plt.title("House Age vs House Price")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(df["Distance to the nearest MRT station"],
            df["House price of unit area"])

plt.xlabel("Distance to MRT")
plt.ylabel("House Price")
plt.title("Distance to MRT vs House Price")
plt.show()
```

```python
#Histogram (Distribution)
plt.hist(df["House price of unit area"], bins=30)
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.title("Distribution of House Prices")
plt.show()
```

```python
#Subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,5))

axes[0].scatter(df["House age"], df["House price of unit area"])
axes[0].set_title("Age vs Price")

axes[1].scatter(df["Distance to the nearest MRT station"],
                df["House price of unit area"])
axes[1].set_title("Distance vs Price")

plt.tight_layout()
plt.show()
```

```python
#Customized Plot (Colors, linewidths)
plt.figure(figsize=(8,5))
plt.plot(df["House age"],
         df["House price of unit area"],
         color="red",
         linewidth=1,
         marker='o')

plt.xlabel("House Age")
plt.ylabel("House Price")
plt.title("Customized Plot")
plt.show()
```

```python
#Seaborn Visualizations: Distribution Plot
import seaborn as sns

sns.histplot(df["House price of unit area"], kde=True)
plt.title("Distribution of House Prices")
plt.show()
```

```python
#Joint Plot
sns.jointplot(x="Distance to the nearest MRT station",
              y="House price of unit area",
              data=df,
              kind="scatter")
plt.show()
```

```python
#Pairplot
sns.pairplot(df)
plt.show()
```

```python
#Boxplot
sns.boxplot(x=df["Number of convenience stores"],
            y=df["House price of unit area"])

plt.title("Convenience Stores vs House Price")
plt.show()
```

```python
#Violin Plot
sns.violinplot(x=df["Number of convenience stores"],
               y=df["House price of unit area"])

plt.show()
```

```python
#Heatmap Correlation Matrix
plt.figure(figsize=(8,6))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
